```python
def saveChunk(data, chunkSize, path, predType):
    n = len(data)
    i = 0
    idx = 0
    while(i*chunkSize < n):
        print(i)
        sub = data[int(i*chunkSize):int((i+1)*chunkSize)].reset_index(drop=True)
        sub.to_parquet(path + f'{predType}_{idx}.pqt')
        i += 1
        idx += 1
```
